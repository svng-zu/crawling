```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
tv= pd.read_csv("C:/Users/PC/OneDrive/바탕 화면/VOD 데이터 크롤링/11_15미확인.csv", encoding ='cp949')
# drama= pd.read_csv("C:/Users/PC/OneDrive/바탕 화면/VOD 데이터 크롤링/11_14영화미확인.csv", encoding ='cp949')
# tv = pd.concat([tv, drama])
tv = tv.replace('  ', ' ', regex =True)
tv_list= list(tv['0'])
tv_list
```

```python
driver = webdriver.Chrome()

image_path = "https://www.wavve.com/"
new_list = []
no_list = []
for i in range(0, len(tv_list)) :
    name = tv_list[i]
    url = f'https://www.wavve.com/search?searchWord={name}'
    try :
        driver.get(url)
        time.sleep(1)
        element = driver.find_element(By.XPATH, '//*[@id="multisection_index_undefined"]/div[2]/div[1]/div/div/a/div[1]/div[2]/picture/img')
        # srcset 속성에서 포스터 URL 추출
        print(element)
        src_attribute = element.get_attribute('src')
        print(src_attribute)
        # 추출한 src 속성에 정규식 적용



        element= driver.find_element(By.XPATH, '//*[@id="multisection_index_undefined"]/div[2]/div[1]/div/div/a')
        element.click()
        time.sleep(1)
        try :
            age_elements = driver.find_elements(By.XPATH, '//*[@id="contents"]/section/div/div/div[2]/dl/dd[2]/span')
            
             #age 요소가 있는 경우
            age = age_elements[0].text
            print(age)
        except :
            #age 요소가 없는 경우
            src = driver.find_element(By.CSS_SELECTOR, '#contents > section > div > div > div > div.video-detail-area > dl > dd:nth-child(4) > span > img')
            age = src.get_attribute('src')
            matches = re.findall(r'\w+-years', age)
            
            age = matches[0]
            age = age.replace('-years', '세 관람가')
            print(age)
            #//*[@id="contents"]/section/div/div/div[2]/dl/dd[2]/span/img
            #contents > section > div > div > div > div.video-detail-area > dl > dd:nth-child(4) > span > img
        try :
            genre = driver.find_element(By.XPATH, '//*[@id="contents"]/section/div/div/div[2]/dl/dd[3]')

        except :
            genre = driver.find_element(By.XPATH, '//*[@id="contents"]/section/div/div/div/div[2]/dl/dd[4]')
            
            
            
        print(genre)
        age = age
        genre = genre.text[4:]

        print(age, genre)
        
        new_list.append([name, genre, age, src_attribute])
    except :
        print(f'{name}오류 발생, 노가다 필요')
        no_list.append(name)
        time.sleep(1)
        pass
```
