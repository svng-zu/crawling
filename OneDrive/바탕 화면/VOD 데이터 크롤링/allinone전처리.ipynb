## 다음 영화 크롤링

```python
import time
import re
driver = webdriver.Chrome()

daum_list = list(merged_df['program_name'])
pro_id = list(merged_df['program_id'])
new_list = []
no_list = []


for i in range(0,len(daum_list)) :

    try :
        name= daum_list[i]
        url = f'https://movie.daum.net/search?q={name}#tab=all'  # 대상 웹페이지의 URL로 변경해야 합니다
        driver.get(url)

        time.sleep(1)

#         xpath = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/span[1]'
        try :
            element = driver.find_element(By.CSS_SELECTOR, 'span.bg_img')

        # 요소의 style 속성 가져오기
            style_attribute = element.get_attribute('style')

        # 정규표현식을 사용하여 URL 추출
            url_pattern = re.compile(r'url\((.*?)\)')
            matches = url_pattern.search(style_attribute)

            if matches:
                poster_url = matches.group(1)
                print(poster_url)
        except :
            poster_url = ''
            pass

        xpath1 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[1]/dd'
        element1 = driver.find_element(By.XPATH, xpath1)

        year = element1.text
        xpath2 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[2]/dd'

        element2 = driver.find_element(By.XPATH, xpath2)
        genre = element2.text

        xpath3 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[3]/dd'
        element3 = driver.find_element(By.XPATH, xpath3)

        age_runtime = element3.text

        new_list.append([pro_id[i], name, year, genre, age_runtime.split(', '), poster_url])
        print('good')
        time.sleep(1)

    except :
        print(f'{name} 은 다시 찾아야 함')
        no_list.append([name, pro_id[i]])
        time.sleep(3)
        pass
```

```python
daum_data = pd.DataFrame(new_list, columns=['program_id', 'program_name', 'release_date', 'genre','age_runtime', 'poster_url'])
daum_data
```

```python
daum_data['poster_url'] = daum_data['poster_url'].replace('"','' , regex=True)
daum_data
```

```python
daum_data.to_csv('./다음크롤링.csv', encoding='cp949', index=False)
```
