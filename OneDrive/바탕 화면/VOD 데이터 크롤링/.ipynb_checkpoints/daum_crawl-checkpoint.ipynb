```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
def get_movie_info(name):
    url = f'https://movie.daum.net/search?q={name}#tab=all'

    # 웹페이지에 요청을 보내고 HTML을 가져옵니다.
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # 영화 정보가 들어있는 부분을 찾습니다.
    movie_info = soup.find('div', class_='info_movie')

    # 필요한 정보를 추출합니다.
    title = movie_info.find('strong', class_='tit_item').text.strip()
    genre = movie_info.find('dd', class_='genre').text.strip()

    print(f'Title: {title}')
    print(f'Genre: {genre}')
    print(f'Director: {director}')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 웹 드라이버 실행
driver = webdriver.Chrome()
name="가디언즈 오브 갤럭시2"
url = f'https://movie.daum.net/search?q={name}#tab=all'  # 대상 웹페이지의 URL로 변경해야 합니다
driver.get(url)

# XPath로 특정 요소를 찾아 텍스트 가져오기
xpath = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[1]/dd'
element1 = driver.find_element(By.XPATH, xpath)
year = element1.text
xpath = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[2]/dd'
element2 = driver.find_element(By.XPATH, xpath)
genre = element2.text
xpath = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[3]/dd'
element3 = driver.find_element(By.XPATH, xpath)
age_runtime = element3.text

# 결과 출력
print(year)
print(genre)
print(age_runtime)
```

```python
import time
import re
driver = webdriver.Chrome()

new_list = []
no_list = []


for i in range(0,len(mv_list)) :
    
    try :
        name= mv_list[i]
        url = f'https://movie.daum.net/search?q={name}#tab=all'  # 대상 웹페이지의 URL로 변경해야 합니다
        driver.get(url)
    
        time.sleep(1)
        

        try :
            element = driver.find_element(By.CSS_SELECTOR, 'span.bg_img')

        # 요소의 style 속성 가져오기
            style_attribute = element.get_attribute('style')

        # 정규표현식을 사용하여 URL 추출
            url_pattern = re.compile(r'url\((.*?)\)')
            matches = url_pattern.search(style_attribute)

            if matches:
                poster_url = matches.group(1)
                print(poster_url)
        except :
            poster_url = 'none'
        
        xpath1 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[1]/dd'
        element1 = driver.find_element(By.XPATH, xpath1)
        year = element1.text
        xpath2 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[2]/dd'
        
        element2 = driver.find_element(By.XPATH, xpath2)
        genre = element2.text
    
        xpath3 = '//*[@id="mainContent"]/div/div[2]/div[2]/ul/li/div/div/dl[3]/dd'
        element3 = driver.find_element(By.XPATH, xpath3)
        
        age_runtime = element3.text
        
        new_list.append([name, year, genre, age_runtime.split(', '), poster_url])
        print('good')
        time.sleep(1)
    
    except : 
        print(f'{name} 은 다시 찾아야 함')
        no_list.append(name)
        time.sleep(3)
        pass
    
```
